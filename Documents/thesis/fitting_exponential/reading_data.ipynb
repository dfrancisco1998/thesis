```python
import numpy as np
import scipy.optimize
iimport matplotlib.pyplot as plt

def find_peaks_window(x, y, window=1):
    peak_x = []
    peak_vals = []
    for i in range(len(y)):
        if i < window:
            continue
        if i > len(y) - window:
            continue
        left_side = y[i-window:i]
        right_side = y[i+1:i+1+window]
        if all(left_side < y[i]) and all(right_side < y[i]):
            peak_x.append(x[i])
            peak_vals.append(y[i])
    return np.array(peak_x), np.array(peak_vals)

def monoExp(x, m, t, b):
    return m * np.exp(-t * x) + b

def decide_peaks(t, ys):
    wind = 3
    limit = 0
    while True and limit < 10: 
        noisy_peak_x, noisy_peak_y = find_peaks_window(t, ys, window=wind)
        if len(noisy_peak_y) <= 10:
            return noisy_peak_x, noisy_peak_y
        wind += 1
        limit += 1

def local_max_peaks(t, ys, peaks, threshold=2):
    y_spl = UnivariateSpline(t,ys,s=0,k=4)
    y_spl_2d = y_spl.derivative(n=1)
    first_ddx = y_spl_2d(t)
    new_peaks = []
    for i in range(len(peaks)):
        # print(peaks[i])
        left = first_ddx[peaks[i] - thresh:peaks[i]]
        # print(left)
        right = first_ddx[peaks[i]: peaks[i] + thresh]
        # print(right)
        if np.average(left) > 0 and np.average(right) < 0: 
            new_peaks.append(peaks[i])

    return new_peaks
    

def fit_T2(ind):
    ys = (T2_data[ind][:, 0] - T2_data[ind][:, 1]) / (T2_data[ind][:, 0] + T2_data[ind][:, 1])
    ti = np.arange(len(ys))
    noisy_peak_x, noisy_peak_y = decide_peaks(ti, ys)
    noisy_peak_x = np.asarray(local_max_peaks(ti, ys, noisy_peak_x, threshold=2))
    noisy_peak_y = ys[noisy_peak_x]
    plt.plot(ti, ys)
    plt.plot(noisy_peak_x, noisy_peak_y, '.', label="data")

    # perform the fit
    p0 = (1.,1.e-1,1.) # start with values near those we expect
    try:
        params, cv = scipy.optimize.curve_fit(monoExp, noisy_peak_x, noisy_peak_y, p0=p0)

    except RuntimeError:
        plt.show()
        return 0
    
    m, t, b = params
    
    # plot the results
    # determine quality of the fit
    squaredDiffs = np.square(noisy_peak_y - monoExp(noisy_peak_x, m, t, b))
    squaredDiffsFromMean = np.square(noisy_peak_y - np.mean(noisy_peak_y))
    rSquared = 1 - np.sum(squaredDiffs) / np.sum(squaredDiffsFromMean)
    
    plt.plot(noisy_peak_x, monoExp(noisy_peak_x, m, t, b), '--', label="fitted")
    plt.title(f"ind: {ind} R = {rSquared}")
    plt.show()
    
    return rSquared

r2 = []
for i in range(len(T2_data)):
    
    r2.append(fit_T2(i))

r2 = np.asarray(r2)


```

```python
import numpy as np
import random as rdm
def genereate_T2star_data(n_samples = 101, noise_lvl = 0.03 , max_stretch = 0.2,lin_log = 1):
    #linspace or logspace between 0 and 1
    if lin_log:
        x_data = (np.linspace(0,1,n_samples))
    else:
        x_data = (np.logspace(-1,0,n_samples))
        x_data = (x_data-np.min(x_data))/(np.max(x_data)-np.min(x_data))

    # generate random T1 value
    T2 = rdm.random()*0.5

    # add some stretching
    T2_stretch = 1+(rdm.random()-0.5)*max_stretch

    # create decay 
    T2_data = np.exp(-(x_data/T2)**T2_stretch) 
    
    # add 13C bumps
    bump_spacing = rdm.random()*0.1+0.1
    T2_data = np.multiply( np.exp(-(x_data/T2)**T2_stretch), np.cos(2*np.pi/bump_spacing*x_data) )

    # add noise
    T2_data = T2_data + np.random.normal(0,noise_lvl,n_samples)

    return x_data, T2_data

```
